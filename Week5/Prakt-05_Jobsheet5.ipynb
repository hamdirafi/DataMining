```python
import pandas as pd
restaurant = pd.read_csv("Restaurant.csv")
restaurant.head()
```

```python
restaurant = restaurant.replace("Yes", True)
restaurant = restaurant.replace("No", False)

restaurant = pd.get_dummies(restaurant)
```

```python
from mlxtend.frequent_patterns import fpgrowth
from mlxtend.frequent_patterns import association_rules

frequent_rest = fpgrowth(restaurant, min_support=0.05, use_colnames=True)
```

```python
rules_rest = association_rules(frequent_rest, metric="confidence", min_threshold=0.5)
rules_rest.sort_values('lift', ascending = False)
```
